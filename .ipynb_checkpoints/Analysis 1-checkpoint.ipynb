```python
import csv
reader = csv.reader(open("data2.csv", "rb"), 
                    delimiter=",", quoting=csv.QUOTE_NONE)
positives = []
negatives = []
for row, record in enumerate(reader):
    fields = []
    i = 1
    I = len(record)
    while i < I:
        fields.append(float(record[i]))
        i = i +1
    if record[0] == "1":
        positives.append(fields)
    elif record[0] == "0":
        negatives.append(fields)
```
